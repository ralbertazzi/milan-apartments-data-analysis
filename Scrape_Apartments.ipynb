```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import urllib
from bs4 import BeautifulSoup
from requests import get
```

```python
def retrieve_annuncio(href):
    soup = BeautifulSoup(get(href).text, 'html.parser')
    address = soup.find('span', class_='im-address__content js-map-address')
    print(address.text.replace('\n', '').strip())
    
def get_price(item):
    if not 'lif__pricing' in item['class']:
        return None
    
    return int(item.text.strip().split('\n')[0].replace('.', '')[2:])

def get_space(items):
    for item in items:
        lif_text = item.find('div', class_='lif__text')
        if lif_text is not None and lif_text.text == 'superficie':
            span = item.find('div', class_='lif__data').find('span')
            return int(span.text)
    return None
```

```python
data = pd.DataFrame(columns=['href', 'address', 'price', 'space'])

BASE_SEARCH_URL = 'https://www.immobiliare.it/affitto-appartamenti/milano/'
for page in range(1000):
    print(page)
    try:
        url = BASE_SEARCH_URL + '?pag=' + str(page)
        response = get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        annunci = soup.find('ul', class_='annunci-list').find_all('li', class_='listing-item premium js-row-detail')
        for annuncio in annunci:
            
            try:       
                features = annuncio.find('ul', class_='listing-features').find_all('li', class_='lif__item')
                price = get_price(features[0])
                space = get_space(features)

                link = annuncio.find('p', class_='titolo text-primary').find('a')
                title, href = link['title'], link['href']
                address = title.split(' ', 1)[1]

                data = data.append({'href': href, 'address': address, 'price': price, 'space': space}, ignore_index=True)
            except Exception as e:
                print(e)
                continue
                
    except Exception as e:
        print(e)
        break
        
data.to_csv('affitti_milano.csv')
```
